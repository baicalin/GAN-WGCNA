```python
from __future__ import absolute_import
from __future__ import division
from __future__ import print_function

import argparse
import re
import sys

import numpy as np

from tensorflow.python import pywrap_tensorflow
from tensorflow.python.platform import app
from tensorflow.python.platform import flags

FLAGS = None
def print_tensors_in_checkpoint_file(file_name, tensor_name, all_tensors,
                                     all_tensor_names=False,
                                     count_exclude_pattern=""):
  """Prints tensors in a checkpoint file.
  If no `tensor_name` is provided, prints the tensor names and shapes
  in the checkpoint file.
  If `tensor_name` is provided, prints the content of the tensor.
  Args:
    file_name: Name of the checkpoint file.
    tensor_name: Name of the tensor in the checkpoint file to print.
    all_tensors: Boolean indicating whether to print all tensors.
    all_tensor_names: Boolean indicating whether to print all tensor names.
    count_exclude_pattern: Regex string, pattern to exclude tensors when count.
  """
  try:
    reader = pywrap_tensorflow.NewCheckpointReader(file_name)
    if all_tensors or all_tensor_names:
      var_to_shape_map = reader.get_variable_to_shape_map()
      for key in sorted(var_to_shape_map):
        print("tensor_name: ", key)
        if all_tensors:
          print(reader.get_tensor(key))
    elif not tensor_name:
      print(reader.debug_string().decode("utf-8"))
    else:
      print("tensor_name: ", tensor_name)
      print(reader.get_tensor(tensor_name))
      disc_dense2 = reader.get_tensor(tensor_name)

    # Count total number of parameters
    print("# Total number of params: %d" % _count_total_params(
        reader, count_exclude_pattern=count_exclude_pattern))
  except Exception as e:  # pylint: disable=broad-except
    print(str(e))
    if "corrupted compressed block contents" in str(e):
      print("It's likely that your checkpoint file has been compressed "
            "with SNAPPY.")
    if ("Data loss" in str(e) and
        any(e in file_name for e in [".index", ".meta", ".data"])):
      proposed_file = ".".join(file_name.split(".")[0:-1])
      v2_file_error_template = """
It's likely that this is a V2 checkpoint and you need to provide the filename
*prefix*.  Try removing the '.' and extension.  Try:
inspect checkpoint --file_name = {}"""
      print(v2_file_error_template.format(proposed_file))
```

```python
from __future__ import division
import math
import random
import pprint
import scipy.misc
import numpy as np
from time import gmtime, strftime
from six.moves import xrange
import matplotlib.pyplot as plt
import os, gzip

from numpy import genfromtxt
from sklearn.preprocessing import MinMaxScaler
import pandas as pd

import time
import pandas as pd
import numpy as np
from sklearn.manifold import TSNE
from matplotlib.lines import Line2D
from MulticoreTSNE import MulticoreTSNE as TSNE

def draw_tsne(df, colar_arr):

    np.random.seed(3)
    n_sne = 5000
    time_start = time.time()
    # tsne = TSNE(n_components=2)
    tsne = TSNE(n_jobs=16)
    tsne_results = tsne.fit_transform(df)

    legend_elems = []

    for key in color_hash:
        legend_elems.append(Line2D([0], [0], marker='o', color='w', label=key,
                              markerfacecolor=color_hash[key], markersize=8))
    
    # equivalent with figure 2
    plt.figure(figsize=(13,9.5))
    plt.scatter(tsne_results[:,0], tsne_results[:,1], color=colar_arr)
    ax = plt.subplot(111)
    ax.legend(handles=legend_elems)
    plt.show()
    return tsne_results
```

```python
def draw_result_map(X, y, model, cols=['',''], title=''):
    # https://jfun.tistory.com/111 참조
    scale = 300
    xmax = X[:,0].max()+1
    xmin = X[:,0].min()-1
    ymax = X[:,1].max()+1
    ymin = X[:,1].min()-1

    xx = np.linspace(xmin,xmax,scale)
    yy = np.linspace(ymin,ymax,scale)
    data1, data2 = np.meshgrid(xx,yy)
    X_grid = np.c_[data1.ravel(), data2.ravel()]
    
    pred_y = model.predict(X_grid)

    fig=plt.figure(figsize=[12,10])

    CS = plt.imshow(pred_y.reshape(scale,scale), interpolation=None, origin='lower',
                    extent=[xmin,xmax,ymin,ymax], alpha=0.3, cmap='gray_r')

    # draw X
    plt.scatter(X[:,0], X[:,1], c=y, s=60)

    plt.xlabel(cols[0])
    plt.ylabel(cols[1])
    plt.colorbar(CS, shrink=0.3)
    plt.title(title,fontsize=20)
    

def svm_to_seperate_SN_CN(tsne_result):
    from sklearn import svm

    class_arr = []

    for col in index_arr:
        class_arr.append(col.split("][")[1])

    class_arr_svm = []
    for cla in class_arr:
        if cla == "CN":
            class_arr_svm.append(0)
        elif cla == "SN":
            class_arr_svm.append(1)
        else:
            class_arr_svm.append(2)


    X = tsne_result[:original_data_len]
    y = class_arr_svm
    clf = svm.SVC(kernel="linear")
#     clf = svm.SVC(kernel="poly")
    clf.fit(X, y)
    draw_result_map(X, y, clf)
    
    predict_result = clf.predict(tsne_result)
    
    denoted_arr = []

    color_hash_etc = {
        0: (1, 0, 0, 0.5),
        1: (0, 1, 0, 0.5),
        2: (0, 0, 1, 0.5)
    }

    for cla in predict_result[-1 * sample_num:]:
        denoted_arr.append(color_hash_etc[cla])
        
    plt.figure(figsize=(13,9.5))
    ax = plt.subplot(111)
    plt.scatter(tsne_result[:,0][-1 * sample_num:], tsne_result[:,1][-1 * sample_num:], color=denoted_arr)
#     ax.legend(handles=legend_elems)

    plt.title("sorting_with_SVM", fontsize=25)
    plt.xlabel('t-SNE 1')
    plt.ylabel('t-SNE 2')
    plt.show()
    
    return predict_result
```

```python
# Repeat for each region and epoch
# BLA, PFC, CPU, VTA, HIP, NAC
region = "NAC"
epoch = 50000
```

```python


# read tensors
reader = pywrap_tensorflow.NewCheckpointReader(r"../GAN_training/checkpoint/WGAN_GP_%s_32_100/WGAN_GP/WGAN_GP.model-%d" % (region, epoch))
gen_dense1 = reader.get_tensor("generator/gen_dense1/kernel")
gen_dense2 = reader.get_tensor("generator/gen_dense2/kernel")
gen_dense1_bias = reader.get_tensor("generator/gen_dense1/bias")
gen_dense2_bias = reader.get_tensor("generator/gen_dense2/bias")
gen_output = reader.get_tensor('generator/gen_output/kernel')
gen_output_bias = reader.get_tensor('generator/gen_output/bias')

# read data
df = pd.read_csv(r'../GAN_training/data/%s_/%s_aug.csv' % (region, region), index_col = 0)
df_transposed = df.transpose()
df_transposed_with_header = df_transposed

color_hash = {
    'SS': (1, 0, 0, 0.5),
    'SC': (0, 1, 0, 0.5),
    'SN': (0, 0, 1, 0.5),
    'CS': (0, 1, 1, 0.5),
    'CC': (1, 0, 1, 0.5),
    'CN': (1, 1, 0, 0.5)
}



col_arr = []
index_arr = df_transposed_with_header.index

for col in index_arr:
    col_arr.append(color_hash[col.split("][")[1]])
original_data_len = len(col_arr)

origin_df = df_transposed
draw_tsne(origin_df, col_arr)
```

```python
# generate fakes

sample_num = 20000
np.random.seed(16)
data_max_value = np.amax(df.values[0, :])
temp_norm = np.random.normal(0.0, data_max_value/10, size=(sample_num, 100))
temp_poisson = np.random.poisson(1, size=(sample_num, 100))
batch_z = np.abs(temp_norm + temp_poisson)

val_arr = batch_z # * 2 
dense1_output =np.matmul(val_arr, gen_dense1) + gen_dense1_bias # input layer 통과
dense1_output_with_lrelu = np.where(dense1_output > 0, dense1_output, dense1_output * 0.2 ) #  leaky Relu 통과
dense2_output = np.matmul(dense1_output_with_lrelu, gen_dense2 ) + gen_dense2_bias # hidden layer 통과
dense2_output_with_lrelu = np.where(dense2_output > 0, dense2_output, dense2_output* 0.2 ) # leaky Relu 통과
gen_output_result = np.matmul(dense2_output_with_lrelu, gen_output ) + gen_output_bias


scaled_df = df_transposed
scaler = MinMaxScaler()
scaler.fit(scaled_df)
scaled_df = scaler.transform(scaled_df)

scaled_df = pd.DataFrame(scaled_df)
# df.columns = predict_df.columns
merged_df = "NaN"
predict_df =pd.DataFrame(gen_output_result)
# merged_df = df_transposed.append( predict_df)
merged_df = pd.concat([scaled_df, predict_df]) 
merged_df.index = range(0, len(merged_df))
a_merged_df = merged_df

agumented_predict_df = predict_df
predicted_col_arr = []

for i in range(sample_num):
    predicted_col_arr.append((0,0,0,0.5))
    
merged_col_arr = col_arr + predicted_col_arr
tsne_result = draw_tsne(a_merged_df.values, merged_col_arr)
```

```python
group_CN, group_SN, group_CN_fake, group_SN_fake = [], [], [], []

for i, v in enumerate(predicted_result[:original_data_len]):
    if v == 0:
        group_CN.append(i)
    elif v == 1:
        group_SN.append(i)
        
for i, v in enumerate(predicted_result[-1 * sample_num:]):
    if v == 0:
        group_CN_fake.append(i + original_data_len)
    elif v == 1:
        group_SN_fake.append(i + original_data_len)
    
```

```python
from tqdm.notebook import tqdm

df_real_SN_id
df_real_CN_id

resembled_SN_vector, resembled_CN_vector = [], [] 
resembled_SN_vector_corr, resembled_CN_vector_corr = [], [] 
print("SN:")
for real_id in tqdm(df_real_SN_id):
    tmp_vectors = {}
    for fake_id in group_SN_fake:
        corr = scaled_df.iloc[real_id,:].corr(a_merged_df.iloc[fake_id,:])
        resembled_SN_vector_corr.append(corr)
#         print(corr)
#         if 0.93 < corr < 0.98:
        if 0.95 < corr:
            tmp_vectors[corr] = batch_z[fake_id-original_data_len]
    if tmp_vectors:
        tmp_vectors = sorted(tmp_vectors.items(), reverse = True)
        resembled_SN_vector.append(np.mean(np.array(tmp_vectors)[0:10, 1], axis=0))
    else:
        print("no match in real_id %d !" % real_id)

print("SN:")
for real_id in tqdm(df_real_CN_id):
    tmp_vectors = {}
    for fake_id in group_CN_fake:
        corr = scaled_df.iloc[real_id,:].corr(a_merged_df.iloc[fake_id,:])
        resembled_CN_vector_corr.append(corr)
#         print(corr)
#         if 0.93 < corr < 0.98:
        if 0.95 < corr:
            tmp_vectors[corr] = batch_z[fake_id-original_data_len]
    if tmp_vectors:
        tmp_vectors = sorted(tmp_vectors.items(), reverse = True)
        resembled_CN_vector.append(np.mean(np.array(tmp_vectors)[0:10, 1], axis=0))
    else:
        print("no match in real_id %d !" % real_id)

```

```python

from tqdm.notebook import tqdm

sim_means = []
# for i in tqdm(range(0, len(resembled_SN_vector))):
# for i in tqdm(range(0, 6)):

latent_vectors_interpolated = []

for sn in tqdm(resembled_SN_vector):
    for cn in resembled_CN_vector:
        small_delta = (cn - sn) / 100.0
        tmp_arr = []

        for i in range(100):
            moved = sn + (small_delta * i)
            tmp_arr.append(moved)
        latent_vectors_interpolated.append(tmp_arr)



gex_sims = []



for moved_vector in tqdm(latent_vectors_interpolated):
    val_arr = moved_vector # * 2 
    dense1_output =np.matmul(val_arr, gen_dense1) + gen_dense1_bias # input layer 통과
    dense1_output_with_lrelu = np.where(dense1_output > 0, dense1_output, dense1_output * 0.2 ) #  leaky Relu 통과
    dense2_output = np.matmul(dense1_output_with_lrelu, gen_dense2 ) + gen_dense2_bias # hidden layer 통과
    dense2_output_with_lrelu = np.where(dense2_output > 0, dense2_output, dense2_output* 0.2 ) # leaky Relu 통과
    gen_output_result = np.matmul(dense2_output_with_lrelu, gen_output ) + gen_output_bias
    gex_sims.append(gen_output_result)

sim_mean = np.mean(gex_sims, axis=0)
sim_means.append(sim_mean)
    
sim_means_mean = np.mean(sim_means, axis=0)
plt.figure(figsize=(13,9.5))
plt.ylim(0, 1.0)
for i in range(0, 50):
    plt.plot(sim_means_mean[:,i])
plt.title("%s-Meas_Mean_gex_0th_to_50" % region, fontsize=25)
plt.xlabel('simulation steps',fontsize=15)
plt.ylabel('gene expression level', fontsize=15)
```
