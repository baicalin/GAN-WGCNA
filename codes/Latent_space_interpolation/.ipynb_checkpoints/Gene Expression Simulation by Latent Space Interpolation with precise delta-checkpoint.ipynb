```python
from __future__ import absolute_import
from __future__ import division
from __future__ import print_function

import argparse
import re
import sys

import numpy as np

from tensorflow.python import pywrap_tensorflow
from tensorflow.python.platform import app
from tensorflow.python.platform import flags

FLAGS = None
def print_tensors_in_checkpoint_file(file_name, tensor_name, all_tensors,
                                     all_tensor_names=False,
                                     count_exclude_pattern=""):
  """Prints tensors in a checkpoint file.
  If no `tensor_name` is provided, prints the tensor names and shapes
  in the checkpoint file.
  If `tensor_name` is provided, prints the content of the tensor.
  Args:
    file_name: Name of the checkpoint file.
    tensor_name: Name of the tensor in the checkpoint file to print.
    all_tensors: Boolean indicating whether to print all tensors.
    all_tensor_names: Boolean indicating whether to print all tensor names.
    count_exclude_pattern: Regex string, pattern to exclude tensors when count.
  """
  try:
    reader = pywrap_tensorflow.NewCheckpointReader(file_name)
    if all_tensors or all_tensor_names:
      var_to_shape_map = reader.get_variable_to_shape_map()
      for key in sorted(var_to_shape_map):
        print("tensor_name: ", key)
        if all_tensors:
          print(reader.get_tensor(key))
    elif not tensor_name:
      print(reader.debug_string().decode("utf-8"))
    else:
      print("tensor_name: ", tensor_name)
      print(reader.get_tensor(tensor_name))
      disc_dense2 = reader.get_tensor(tensor_name)

    # Count total number of parameters
    print("# Total number of params: %d" % _count_total_params(
        reader, count_exclude_pattern=count_exclude_pattern))
  except Exception as e:  # pylint: disable=broad-except
    print(str(e))
    if "corrupted compressed block contents" in str(e):
      print("It's likely that your checkpoint file has been compressed "
            "with SNAPPY.")
    if ("Data loss" in str(e) and
        any(e in file_name for e in [".index", ".meta", ".data"])):
      proposed_file = ".".join(file_name.split(".")[0:-1])
      v2_file_error_template = """
It's likely that this is a V2 checkpoint and you need to provide the filename
*prefix*.  Try removing the '.' and extension.  Try:
inspect checkpoint --file_name = {}"""
      print(v2_file_error_template.format(proposed_file))
reader = pywrap_tensorflow.NewCheckpointReader(r"../Cocaine-GAN/checkpoint/WGAN_GP_BLA_10k_32_100/WGAN_GP/WGAN_GP.model-20000")
gen_dense1 = reader.get_tensor("generator/gen_dense1/kernel")
gen_dense2 = reader.get_tensor("generator/gen_dense2/kernel")
gen_dense1_bias = reader.get_tensor("generator/gen_dense1/bias")
gen_dense2_bias = reader.get_tensor("generator/gen_dense2/bias")
gen_output = reader.get_tensor('generator/gen_output/kernel')
gen_output_bias = reader.get_tensor('generator/gen_output/bias')
```

```python
from __future__ import division
import math
import random
import pprint
import scipy.misc
import numpy as np
from time import gmtime, strftime
from six.moves import xrange
import matplotlib.pyplot as plt
import os, gzip

from numpy import genfromtxt
from sklearn.preprocessing import MinMaxScaler
import pandas as pd

epoch = 40000
region = "BLA"
df = pd.read_csv(r'../Cocaine-GAN/data/%s_10k/%s-DEG_aug.csv' % (region, region), index_col = 0)
df_transposed = df.transpose()
df_transposed_with_header = df_transposed


color_hash = {
    'SS': (1, 0, 0, 0.5),
    'SC': (0, 1, 0, 0.5),
    'SN': (0, 0, 1, 0.5),
    'CS': (0, 1, 1, 0.5),
    'CC': (1, 0, 1, 0.5),
    'CN': (1, 1, 0, 0.5)
}

col_arr = []
index_arr = df_transposed_with_header.index

for col in index_arr:
    col_arr.append(color_hash[col.split("][")[1]])

import time
import pandas as pd
import numpy as np
from sklearn.manifold import TSNE

np.random.seed(2)
n_sne = 5000
time_start = time.time()
tsne = TSNE(n_components=2)
tsne_results = tsne.fit_transform(df_transposed)

df_tsne = df_transposed.copy()
df_tsne['x-tsne'] = tsne_results[:,0]
df_tsne['y-tsne'] = tsne_results[:,1]

# equivalent with figure 2
plt.figure(figsize=(13,9.5))
plt.scatter(df_tsne['x-tsne'], df_tsne['y-tsne'], color=col_arr)
plt.show()
```

```python
sample_num = 10000
np.random.seed(2)
data_max_value = np.amax(df.values[0, :])
temp_norm = np.random.normal(0.0, data_max_value/10, size=(sample_num, 100))
temp_poisson = np.random.poisson(1, size=(sample_num, 100))
batch_z = np.abs(temp_norm + temp_poisson)
```

```python
val_arr = batch_z # * 2 
dense1_output =np.matmul(val_arr, gen_dense1) + gen_dense1_bias # input layer 
dense1_output_with_lrelu = np.where(dense1_output > 0, dense1_output, dense1_output * 0.2 ) #  leaky Relu 
dense2_output = np.matmul(dense1_output_with_lrelu, gen_dense2 ) + gen_dense2_bias # hidden layer 
dense2_output_with_lrelu = np.where(dense2_output > 0, dense2_output, dense2_output* 0.2 ) # leaky Relu 
gen_output_result = np.matmul(dense2_output_with_lrelu, gen_output ) + gen_output_bias
```
